```python
from ScalableVolumetricBenchmark import *

#tket for testing
from pytket.circuit import Circuit
from pytket.extensions.qiskit import AerBackend
from pytket.backends.backendresult import BackendResult
from pytket.qasm import circuit_from_qasm_str

#for visualization of the circuit 
from pytket.circuit.display import get_circuit_renderer

#import qiskit noisemodel for the simulation
from qiskit_aer.noise import NoiseModel, depolarizing_error, ReadoutError, pauli_error



circuit_renderer = get_circuit_renderer() # Instantiate a circuit renderer
circuit_renderer.set_render_options(zx_style=True) # Configure render options
circuit_renderer.config.render_options.condense_c_bits = False # You can also set the properties on the instance directly
circuit_renderer.config.min_height = "300px" # Change the display height

# import python modules for visualization
import matplotlib.pyplot as plt
```

# Clifford Volume Benchmark with Tket

```python
#pytket accept qasm2 
n_qubits = 4
sample_size = 4
shots = 1024

benchmark = CliffordVolumeBenchmark(
    number_of_qubits=n_qubits,
    sample_size=sample_size,
    format="qasm2",
    target_sdk="tket",
    shots=shots,
    auto_save=False,
)
benchmark.create_benchmark()
benchmark
```

```python
benchmark_circuits = benchmark.get_all_circuits()

#load qasm to pytket
print(benchmark_circuits[0])
circuit_renderer.render_circuit_jupyter(circuit_from_qasm_str(benchmark_circuits[0]))


```
